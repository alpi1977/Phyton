```python
limit = int(input("Please enter a limit number: "))
primelist = []
for i in range(2,limit + 1):
  for j in range(2,limit + 1):
        if i%j == 0 and i != j:
            break
        else:
            for k in primelist:
                if  i%k ==0:
                    break
            else:
                primelist.append(i)

print (primelist)

```
